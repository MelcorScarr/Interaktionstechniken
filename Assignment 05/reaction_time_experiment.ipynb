They should press one of two keys depending on which of the two stimuli is presented. The participants are distracted by a signal noise in half of the tests, whereas there is no distraction in the other half. The test group gets to use a laptop which already runs the needed script and also has to wear headphones as there are audio signals. The test subjects then get informed by the test leader how the testing works. As part of the study we are using a indirect pointing device, as a keyboard is used to relay the chosen option.

### Hypothesis
Our hypothesis is that, where a distraction is used, a higher reaction time can be observed. Thus, our independant and dependant variables are distraction and reaction time respectively.


```python
from pylab import *
import matplotlib
import numpy
import scipy.stats
%matplotlib inline
rcParams['figure.figsize'] = (16,9)
```

# Extracting data from csv

```python
data1 = genfromtxt("reaction1.csv", dtype=None, delimiter=",", names = True, encoding = None)
data2 = genfromtxt("reaction2.csv", dtype=None, delimiter=",", names = True, encoding = None)
data3 = genfromtxt("reaction3.csv", dtype=None, delimiter=",", names = True, encoding = None)
data4 = genfromtxt("reaction4.csv", dtype=None, delimiter=",", names = True, encoding = None)
```

The following line charts show the reaction times measured in the experiment

```python
plot(data1['Reactiontime'], '-r')
xlabel('Trials')
ylabel('Reaction time for participant 1')
```

```python
plot(data2['Reactiontime'], '-r')
xlabel('Trials')
ylabel('Reaction time for participant 2')
```

```python
plot(data3['Reactiontime'], '-r')
xlabel('Summary of Trials')
ylabel('Reaction time for participant 3')
```

```python
plot(data4['Reactiontime'], '-r')
xlabel('Summary of Trials')
ylabel('Reaction time for participant 4')
```

At a first glance, there is not as much fatigue at later trials as one might have expected. Similarily, there does not seem to be much of a learn factor either. Anyway, as a next step we seperate the data by whether a distraction was used or not.

```python
allDistractions = np.concatenate((data1['Distraction'], data2['Distraction'], data3['Distraction'], data4['Distraction']), axis=0)
allAttentives = np.concatenate((data1['Attentive'], data2['Attentive'], data3['Attentive'], data4['Attentive']), axis=0)
allTimes = np.concatenate((data1['Reactiontime'], data2['Reactiontime'], data3['Reactiontime'], data4['Reactiontime']),axis=0)


reactionOnNonAttentive = (np.ma.masked_array(allTimes, allAttentives))
reactionOnUndistracted = (np.ma.masked_array(allTimes, allDistractions))
reactionOnAttentive = (np.ma.masked_array(allTimes, ~allAttentives))
reactionOnDistracted =(np.ma.masked_array(allTimes, ~allDistractions))

print (reactionOnNonAttentive)
print (reactionOnUndistracted)
```

# Scatter Plots
Overall, the scatter plots do not show a significant difference between distracted and non distracted task completion. Since the log files do not include the type of condition (instead they show the stimulus like the shown word or color of the square), no condition based graphs can be drawn. This is a bug in the logging of the experiment. Since the type of the condition can be derived from the Stimulus, we edited our text files accordingly. This is why you save data redundantly in experiments! :)

```python
scatter(range(len(reactionOnUndistracted)), reactionOnUndistracted, color='blue', marker = 'o')
scatter(range(len(reactionOnDistracted)), reactionOnDistracted, color='yellow', marker = 'o')
scatter(range(len(reactionOnAttentive)), reactionOnAttentive, color='red', marker = '.')
scatter(range(len(reactionOnNonAttentive)), reactionOnNonAttentive, color='green', marker = '.')

legend1 = matplotlib.patches.Patch(color='green', label='Non-Attentive Task')
legend2 = matplotlib.patches.Patch(color='blue', label='Undistracted Task')
legend3 = matplotlib.patches.Patch(color='red', label='Attentive Task')
legend4 = matplotlib.patches.Patch(color='yellow', label='Distracted Task')

legend(handles = [legend1, legend2, legend3, legend4])

title('Reaction Time on Task')
xlabel('Task #')
ylabel('Reaction Time')
```

Curiously enough, Undistracted tasks seem to have taken longer in some cases! Whereas there seems to be no indication that the kind of task has a lot of impact. Let's check if this is true statistically.

# t-test

Let's run a t-test to see if our impression from the plots prove right:

```python
scipy.stats.ttest_ind(np.ma.MaskedArray.compressed(reactionOnUndistracted), np.ma.MaskedArray.compressed(reactionOnDistracted), equal_var=False)
```

With a p-value of 0.35, we cannot statistically claim that the two populations were non-identical. Thus, we cannot say that our hypothesis that distracted reaction times are longer than non-distracted ones. This also means that our impression from the scatterplot cannot be confirmed on a statistical basis.

# Conclusion
As seen in the scatter plots, there is a very slight indication that distraction has much impact on response time at first. However, we were not able to confirm this through the student's t-test, as the two populations are not significantly different from each other.
